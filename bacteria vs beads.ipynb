```python
import matplotlib.pyplot as plt
```

```python

samples=[5,10,20]
```

```python
area=[4,15,30]
```

```python
plt.bar(samples,area,width=1,color=['red','green','blue'])
plt.xlabel('grams of garlic in beads') 
plt.ylabel('area of inhibition in E.coli')
plt.title('bacteria vs agrose beads') 

plt.legend        
plt.show
```

```python
x=[8,16,24]
y=[10,9,11]
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=12) 
plt.xlabel('time period')
plt.ylabel('inhibition rate')
plt.title('analyzing sample3 beads') 
```

```python
x1 = [5,8,10] 
y1 = [4,3.3,2] 
# plotting the line 1 points  
plt.plot(x1, y1, label = "growth rate of bacteria") 
  
# line 2 points 
x2 = [5,8,10] 
y2 = [3.2,3.4,3.7] 
# plotting the line 2 points  
plt.plot(x2, y2, label = "inhibition rate with beads") 
  
# naming the x axis 
plt.xlabel('area') 
# naming the y axis 
plt.ylabel('rate') 
# giving a title to my graph 
plt.title('multiplication control rate of bacteria on agrose beads') 
plt.legend() 
plt.show
  
```
